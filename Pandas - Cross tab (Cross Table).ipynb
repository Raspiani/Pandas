## Pandas - Cross tab (Cross Table)

Dalam tutorial ini kita akan mempelajari bagaimana membuat cross tab (2 way cross table or  3 way cross table or contingency table). 

Yang akan kita pelajari yaitu:
1. 2 way cross table or contingency table
2. 3 way cross table or contingency table 

```python
#memasukan modul yang digunakan 
import pandas as pd
import numpy as np
```

```python
#membuat dataFrame
df = {'Name':['Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine',
            'Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine'],
      'Exam':['Semester 1','Semester 1','Semester 1','Semester 1','Semester 1','Semester 1',
            'Semester 2','Semester 2','Semester 2','Semester 2','Semester 2','Semester 2'],
      'Subject':['Mathematics','Mathematics','Mathematics','Science','Science','Science',
               'Mathematics','Mathematics','Mathematics','Science','Science','Science'],
      'Result':['Pass','Pass','Fail','Pass','Fail','Pass','Pass','Fail','Fail','Pass','Pass','Fail']}
 

data = pd.DataFrame(df)
data
```

#### 2 way cross table or contingency table

```python
pd.crosstab(data.Subject, data.Result, margins = True)
```

#### 3 way cross table or contingency table

```python
pd.crosstab([data.Subject, data.Exam], data.Result, margins = True)
```
