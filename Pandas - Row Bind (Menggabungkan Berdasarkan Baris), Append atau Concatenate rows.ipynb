<p>Dalam tutorial ini kita akan mempelajari bagaimana menggabungkan baris pada dataFrame dengan fungsi append() dan fungsi concat(). Misalnya cara menggabungkan baris dari dua dataFrame yang ada.</p>

```python
#memasukan modul yang digunakan
import pandas as pd
import numpy as np
```

```python
#membuat dataFrame
df = { 'Name':['Alisa','Bobby','Cathrine','Madonna','Rocky','Sebastian','Jaqluine',
   'Rahul','David'],
       'Score1':[62,47,55,74,31,77,85,63,42],
       'Score2':[89,87,67,55,47,72,76,79,44],
       'Score3':[56,86,77,45,73,62,74,89,71]}

data = pd.DataFrame(df)
data
```

```python
df = { 'Name':['Andrew','Ajay','Teresa'],
       'Score1':[32,71,57],
       'Score2':[92,99,69],
       'Score3':[67,97,68]}

data1 = pd.DataFrame(df)
data1
```

#### Metode 1 : Menggabungkan Baris dari dua dataFrame dengan Fungsi concat()

```python
pd.concat([data, data1])
```

#### Metode 2 : Menggabungkan Baris dari dua dataFrame dengan Fungsi append()

```python
data.append(data1)
```

#### Concatenate atau append baris dari dataFrame dengan nama kolom yang berbeda

```python
df = {
    'Name':['Jack','danny','vishwa'],
   'Score1':[32,71,70],
   'Score4':[72,91,89],
   'Score5':[57,72,78]}

data2 = pd.DataFrame(df)
data2
```

#### Concatenate (Menggabungkan) data dengan data2 yang mana nama kolomnya berbeda menggunakan fungsi concat()

```python
pd.concat([data, data2])
```

#### Concatenate (Menggabungkan) data dengan data2 yang mana nama kolomnya berbeda menggunakan fungsi append()

```python
data.append(data2)
```
