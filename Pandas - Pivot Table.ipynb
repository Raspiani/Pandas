## Pandas - Pivot Tabel

    Berikut ini langkah-langkah untuk membuat pivot tabel dari suatu data. Dimana jenis pivot terdiri dari: 
    1. Pivot Biasa 
    2. Pivot Tabel

### 1. Pivot Biasa

```python
#memasukan modul yang digunakan
import pandas as pd
import numpy as np
```

```python
#memasukan data 
data = pd.read_excel("C:/WinPython_64bit/notebooks/cuaca.xlsx")
data
```

```python
#membuat pivot berdasarkan index='city', columns = 'date'
data.pivot(index = 'city', columns = 'date')
```

```python
data.pivot(index = 'city', columns = 'date', values = 'humidity')
```

```python
data.pivot(index = 'date', columns = 'city')
```

```python
data.pivot(index = 'humidity', columns = 'city')
```

### 2. Pivot Table

```python
data
```

```python
data.pivot_table(index = 'city', columns = 'date')
```

```python
#menambahkan aggfunc = "mean"
#aggfunc = "mean" adalah default
data.pivot_table(index = 'city', columns = 'date', aggfunc = "mean")
```

```python
#menambahkan aggfunc = "count"
data.pivot_table(index = 'city', columns = 'date', aggfunc = "count")
```

#### Margins

```python
#menambahkan  margins = True, dan aggfunc = np.sum
data.pivot_table(index = 'city', columns = 'date', margins = True, aggfunc = np.sum)
```

#### Catatan : 

    - aggfunct dapat diisi sesuai kebutuhannya masing-masing
    - misalnya 
        aggfunc = np.sum
        aggfunc = "mean"
        aggfunc = "diff"
        
Dapat dilihat lebih lengkap di https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html

```python
#memasukan data tambahan
data1 = pd.read_excel("C:/WinPython_64bit/notebooks/cuaca1.xlsx")
data1
```

```python
#untuk memeriksa apakah kolom date sudah bertipe tanggal atau Timestamp  
#jika sudah bertipe timestamp maka tidak usah diubah atau di transformasi menjadi timestamp kembali
type(data1['date'][0])
```

```python
#untuk mentransformasi kolom date yang bukan bertipe date menjadi tipe date
data1['date'] = pd.to_datetime(data1['date'])
data1['date'][0]
```

```python
#membentuk pivot tabel dengan grouper
#index = pd.Grouper(freq = 'M', key = 'date'), columns = 'city'
#freq = 'M' dapat dilihat lebih lengkap di http://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html

data1.pivot_table(index = pd.Grouper(freq = 'M', key = 'date'), columns = 'city')

#hasilnya: kolom dibentuk berdasarkan tanggal terakhir dengan nilai rata-ratanya
```
