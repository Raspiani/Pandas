## Pandas - Statistika Deskriptif

Statistika deskriptif atau biasa disebut dengan summary statistics bisa diperoleh dengan menggunakan fungsi describe().

Fungsi describe() memberikan nilai rata-rata (mean), std, dan IQR Values.

1. Secara umum fungsi describe() mengeluarkan kolom-kolom yang berisi karakter (string) dan memberikan summary statistics dari kolom-kolom yang berisi angka (numerik)
2. Kita perlu menambahkan sebuah variabel yang bernama include = 'all' untuk mendapatkan summary statistics atau statistika deskriptif dari kedua kolom tersebut (kolom yang berisi karakter dan kolom yang berisi numerik)

```python
#memasukan modul yang digunakan 
import pandas as pd
import numpy as np
```

```python
#membuat sebuah series dictionary
df = {'Nama' : pd.Series(['Alisa', 'Bobby', 'Cathrine', 'Madonna','Rocky', 'Sebastian', 'Jaqluine', 'Rahul', 'David', 'Andrew', 'Ajay', 'Teresa']),
      'Umur' : pd.Series([26,27,25,24,31,27,25,33,42,32,51,47]),
      'Nilai' : pd.Series([89,87,67,55,47,72,76,79,44,92,99,69])}

#membuat dataFrame
data = pd.DataFrame(df)
data
```

#### Summary Statistic of the numeric columns

```python
#statistika deskriptif
data.describe()
```

#### Summary Statistic of the character columns

```python
data.describe(include = ['object'])
```

```python
data.describe(include='all')
```
