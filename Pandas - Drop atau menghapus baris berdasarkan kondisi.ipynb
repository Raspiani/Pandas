## Pandas - Drop atau menghapus baris berdasarkan kondisi

Dalam tutorial ini kita akan mempelajari cara drop atau menghapus baris berdasarkan index, menghapus baris berdasarkan kondisi dan menghapus baris berdasarkan posisi.

```python
#memasukan modul yang digunakan 
import  pandas as pd
import numpy as np
```

```python
#membuat dataFrame

df = {'Name':['Alisa','Bobby','jodha','jack','raghu','Cathrine',
            'Alisa','Bobby','kumar','Alisa','Alex','Cathrine'],
      'Age':[26,24,23,22,23,24,26,24,22,23,24,24],
      'Score':[85,63,55,74,31,77,85,63,42,62,89,77]}

data = pd.DataFrame(df, columns = ['Name', 'Age', 'Score'])
data
```

#### Simply drop a row or observation (Hanya menghapus beberapa baris)

```python
data.drop([1,2])
```

#### Drop a row or observation by condition (Menghapus baris berdasarkan kondisi)

```python
data[data.Name != 'Alisa']
```

#### Drop a row or observation by index (Menghapus baris berdasarkan index)

```python
data.drop(data.index[2])
```

#### Drop the row by position (Menghapus baris berdasarkan posisi)

```python
data[:-3]
```
