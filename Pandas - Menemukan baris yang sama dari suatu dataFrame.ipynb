## Pandas - Menemukan baris yang sama dari suatu dataFrame

Dalam tutorial ini kita akan mempelajari cara menemukan baris yang sama dari suatu dataFrame dengan fungsi duplicated().

Kita akan memberikan tanda pada baris sebagai TRUE jika baris tersebut adalah duplikat dan False jika bukan baris duplikat.

```python
#memasukan modul yang digunakan 
import pandas as pd
import numpy as np
```

```python
df = { 'Name':['Alisa','Bobby','jodha','jack','raghu','Cathrine',
            'Alisa','Bobby','kumar','Alisa','Alex','Cathrine'],
       'Age':[26,24,23,22,23,24,26,24,22,23,24,24],
       'Score':[85,63,55,74,31,77,85,63,42,62,89,77]}

data = pd.DataFrame(df, columns=['Name', 'Age', 'Score'])
data
```

#### Menemukan baris yang sama (duplikat)

```python
data['adalah_duplikat'] = data.duplicated()
data
```
