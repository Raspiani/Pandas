## Pandas - Column Bind (Menggabungkan kolom) atau Concatenate columns 

Dalam tutorial ini kita akan mempelajari cara menggabungkan kolom dari suatu dataFrame menggunakan fungsi concat() dengan contoh dataFrame berikut ini. 

Gambaran alur kerja tersebut yaitu:

<p><img alt="" src="8. BELAJAR DATA SCIENCE/column_bind.jpg" /></p>

```python
#memasukan modul yang digunakan 
import pandas as pd
import numpy as np
```

```python
#membuat dataFrame
df = { 'Name':['Alisa','Bobby','Cathrine','Madonna','Rocky','Sebastian','Jaqluine',
'Rahul','David'],
       'Score1':[62,47,55,74,31,77,85,63,42],
       'Score2':[89,87,67,55,47,72,76,79,44]}

data = pd.DataFrame(df)
data
```

```python
#membuat dataFrame baru
df = {'Score3':[56,86,77,45,73,62,74,89,71]}

data1 = pd.DataFrame(df)
data1
```

#### Column Bind atau Concatenate Columns (Menggabungkan Kolom) dari dua dataFrame 

```python
pd.concat([data,data1], axis=1, ignore_index=True)
```

```python
pd.concat([data,data1], axis=1, ignore_index=False)
```
