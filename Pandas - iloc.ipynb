### Pandas - iloc

    Berikut ini cara yang dapat kita gunakan dari fungsi iloc.

```python
#memasukan modul yang digunakan
import pandas as pd
import numpy as np
```

```python
#memasukan data
df = pd.read_csv("C:/Users/Yanie/PT Premier Optima Sattiga/adult.csv")
```

```python
#membaca data
df.head()
```

```python
#melihat jumlah baris yang kosong dan memasukannya hasilnya ke dalam variabel df1
df1 = df.isnull().sum()
```

```python
#mengecek apakah ada missing value atau tidak secara keseluruhan
df2 = df1.isnull().any().any()
df2
```

```python
#untuk menampilkan data dengan baris kelipatan 3 dan kolom kelipatan 3
df.iloc[::2,::2]
```

```python
#untuk menampilkan data dengan baris genap dan kolom ganjil misalnya age, fnlwgt, education-num, occupation dst.
df.iloc[::2, 0::2]
```

```python
#fibonacci_numbers
fibonacci_numbers = [0, 1]
for i in range(50):
    fibonacci_numbers.append(fibonacci_numbers[i]+fibonacci_numbers[i+1])
```

```python
#untuk menampilkan data dengan baris data adalah indeks dari fibonacci_numbers
df.iloc[fibonacci_numbers]
```
